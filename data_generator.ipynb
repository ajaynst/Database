```python
import random
from datetime import date, timedelta
```

```python
random.seed(42)

# ---------- CONFIG ----------
NUM_BRANCHES = 20
NUM_CUSTOMERS = 50
NUM_ACCOUNTS = 50
NUM_LOANS = 40
NUM_REPAYMENTS = 20
```

```python
cities = [
    "Mumbai", "Delhi", "Bangalore", "Chennai", "Hyderabad",
    "Pune", "Kolkata", "Ahmedabad", "Jaipur", "Chandigarh"
]

loan_types = ["HOME", "PERSONAL", "AUTO"]
account_types = ["SAVINGS", "CURRENT", "FD"]
kyc_statuses = ["VERIFIED", "PENDING"]
loan_statuses = ["ACTIVE", "CLOSED", "NPA"]

start_date = date(2020, 1, 1)

def rand_date(start, end_days=1500):
    return start + timedelta(days=random.randint(0, end_days))
```

```python
# ---------- BRANCHES ----------
print("-- Branches")
for i in range(1, NUM_BRANCHES + 1):
    print(
        f"INSERT INTO branches (branch_name, city) VALUES "
        f"('Branch_{i}', '{random.choice(cities)}');"
    )
```

```python
# ---------- CUSTOMERS ----------
print("\n-- Customers")
for i in range(1, NUM_CUSTOMERS + 1):
    print(
        f"INSERT INTO customers (full_name, kyc_status, created_at) VALUES "
        f"('Customer_{i}', '{random.choice(kyc_statuses)}', '{rand_date(start_date)}');"
    )
```

```python
# ---------- DEPOSIT ACCOUNTS ----------
print("\n-- Deposit Accounts")
for i in range(1, NUM_ACCOUNTS + 1):
    print(
        f"INSERT INTO deposit_accounts "
        f"(customer_id, branch_id, account_type, balance, opened_at) VALUES "
        f"({i}, {random.randint(1, NUM_BRANCHES)}, "
        f"'{random.choice(account_types)}', "
        f"{round(random.uniform(1000, 500000), 2)}, "
        f"'{rand_date(start_date)}');"
    )
```

```python
# ---------- LOANS ----------
print("\n-- Loans")
for i in range(1, NUM_LOANS + 1):
    principal = random.randint(50_000, 2_000_000)
    print(
        f"INSERT INTO loans "
        f"(customer_id, branch_id, loan_type, principal, interest_rate, tenure_months, disbursed_at, status) VALUES "
        f"({random.randint(1, NUM_CUSTOMERS)}, "
        f"{random.randint(1, NUM_BRANCHES)}, "
        f"'{random.choice(loan_types)}', "
        f"{principal}, "
        f"{round(random.uniform(8.0, 18.0), 2)}, "
        f"{random.choice([12, 24, 36, 60, 120])}, "
        f"'{rand_date(start_date)}', "
        f"'{random.choice(loan_statuses)}');"
    )
```

```python
# ---------- REPAYMENTS ----------
print("\n-- Repayments")
for i in range(1, NUM_REPAYMENTS + 1):
    due = rand_date(start_date)
    paid = due + timedelta(days=random.choice([0, 5, 15, 30]))
    print(
        f"INSERT INTO repayments "
        f"(loan_id, due_date, paid_date, amount_due, amount_paid) VALUES "
        f"({random.randint(1, NUM_LOANS)}, "
        f"'{due}', "
        f"'{paid}', "
        f"{round(random.uniform(3000, 50000), 2)}, "
        f"{round(random.uniform(2000, 50000), 2)});"
    )
```
